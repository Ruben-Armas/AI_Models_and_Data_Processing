## Datos

### Objetivo

Predecir la variable ordinal **damage_grade**, que representa el nivel de daño del edificio afectado por el terremoto. 

Hay 3 grados de daños:
1. representa un daño bajo
2. representa un daño medio
3. representa una destrucción casi completa

### Introducción al DataSet

El conjunto de datos consiste en información sobre la **estructura** de los edificios y su **propiedad legal**.

**Cada fila** del conjunto de datos representa **un edificio** concreto de la región afectada por el terremoto de Gorkha.

Hay **39 columnas** en este conjunto de datos, donde la columna **building_id** --> **identificador** único y aleatorio.

Las 38 características restantes se describen en la sección siguiente. Las variables categóricas se han ofuscado con caracteres ascii minúsculos aleatorios. La aparición del mismo carácter en columnas distintas no implica el mismo valor original.

### Descripción de cada columna

* **geo_level_1_id, geo_level_2_id, geo_level_3_id** (tipo: int): región geográfica en la que existe el edificio, desde la más grande (nivel 1) hasta la subregión más específica (nivel 3). Valores posibles: nivel 1: 0-30, nivel 2: 0-1427, nivel 3: 0-12567.
* **count_floors_pre_eq** (tipo: int): número de plantas del edificio antes del terremoto.
* **age** (tipo: int): edad del edificio en años.
* **area_percentage** (tipo: int): área normalizada de la huella del edificio.
* **height_percentage** (tipo: int): altura normalizada de la huella del edificio.

* **land_surface_condition** (tipo: categorical): estado de la superficie del terreno donde se construyó el edificio. Valores posibles: n, o, t.
* **foundation_type** (tipo: categórico): tipo de cimentación utilizado durante la construcción. Valores posibles: h, i, r, u, w.
* **roof_type** (tipo: categórico): tipo de tejado utilizado durante la construcción. Valores posibles: n, q, x.
* **ground_floor_type** (tipo: categorical): tipo de la planta baja. Valores posibles: f, m, v, x, z.
* **other_floor_type** (tipo: categorical): tipo de construcciones utilizadas en plantas superiores a la baja (excepto de tejado). Valores posibles: j, q, s, x.
* **position** (tipo: categorical): posición del edificio. Valores posibles: j, o, s, t.
* **plan_configuration** (tipo: categorical): configuración de la planta del edificio. Valores posibles: a, c, d, f, m, n, o, q, s, u.
* **has_superstructure_adobe_mud** (tipo: binario): variable bandera que indica si la superestructura era de Adobe/Mud.
* **has_superstructure_mud_mortar_stone** (tipo: binario): variable de bandera que indica si la superestructura estaba hecha de Barro-Mortero-Piedra.
* **has_superstructure_stone_flag** (tipo: binario): variable bandera que indica si la superestructura era de Piedra.
* **has_superstructure_cement_mortar_stone** (tipo: binario): variable bandera que indica si la superestructura era de Mortero de Cemento - Piedra.
* **has_superstructure_mud_mortar_brick** (tipo: binario): variable de bandera que indica si la superestructura era de Mortero de Barro - Ladrillo.
* **has_superstructure_cement_mortar_brick** (tipo: binario): variable de bandera que indica si la superestructura era de Mortero de Cemento - Ladrillo.
* **has_superstructure_timber** (tipo: binario): variable de bandera que indica si la superestructura es de madera.
* **has_superstructure_bamboo** (tipo: binario): variable de bandera que indica si la superestructura era de Bambú.
* **has_superstructure_rc_non_engineered** (tipo: binario): variable de bandera que indica si la superestructura era de hormigón armado no diseñado.
* **has_superstructure_rc_engineered** (tipo: binario): variable de bandera que indica si la superestructura es de hormigón armado diseñado.
* **has_superstructure_other** (tipo: binario): variable de bandera que indica si la superestructura era de cualquier otro material.
* **legal_ownership_status** (tipo: categorical): estado de propiedad legal del terreno donde se construyó el edificio. Valores posibles: a, r, v, w.
* **count_families** (tipo: int): número de familias que viven en el edificio.
* *has_secondary_use** (tipo: binario): variable bandera que indica si el edificio se utilizó para algún fin secundario.
* **has_secondary_use_agriculture** (tipo: binario): variable flag que indica si el edificio se utilizó con fines agrícolas.
* **has_secondary_use_hotel** (tipo: binario): variable de bandera que indica si el edificio se utilizó como hotel.
* **has_secondary_use_rental** (tipo: binario): variable de bandera que indica si el edificio se utilizó con fines de alquiler.
* **has_secondary_use_institution** (tipo: binario): variable de bandera que indica si el edificio se utilizó como sede de alguna institución.
* **has_secondary_use_school** (tipo: binario): variable de bandera que indica si el edificio se utilizó como escuela.
* **has_secondary_use_industry** (tipo: binario): variable de bandera que indica si el edificio se utilizó con fines industriales.
* **has_secondary_use_health_post** (tipo: binario): variable de bandera que indica si el edificio se utilizó como puesto sanitario.
* **has_secondary_use_gov_office** (tipo: binario): variable de bandera que indica si el edificio se utilizó fas una oficina gubernamental.
* **has_secondary_use_use_police** (tipo: binario): variable de bandera que indica si el edificio se utilizó como comisaría de policía.
* **has_secondary_use_other** (tipo: binario): variable de bandera que indica si el edificio se utilizó secundariamente para otros fines.

Selección de un porcentaje del dataset

## Función para reducir todo el dataFrame de manera estratificada/"proporcional"

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

def reduce_df(df, percentage, stratify_col=None):
    """
    Función para reducir un DataFrame en un porcentaje determinado y mantener la proporción de una columna objetivo (stratify_col)
    
    :param df: DataFrame a reducir
    :param percentage: Porcentaje a reducir del DataFrame (entre 0 y 1)
    :param stratify_col: Columna a estratificar para mantener la proporción (opcional)
    :return: DataFrame reducido
    """
    if percentage <= 0 or percentage >= 1:
        raise ValueError("El porcentaje debe estar entre 0 y 1")
    
    if stratify_col is not None:
        if stratify_col not in df.columns:
            raise ValueError(f"{stratify_col} no se encuentra en el DataFrame")
        stratify_counts = df[stratify_col].value_counts()
        n_samples = int(df.shape[0] * percentage)
        n_samples_per_class = (stratify_counts * percentage).astype(int)
        n_samples_per_class = n_samples_per_class.clip(1, None)
        reduced_df = pd.concat([df[df[stratify_col] == value].sample(n_samples_per_class[value])
                                for value in stratify_counts.index])
    else:
        n_samples = int(df.shape[0] * percentage)
        reduced_df = df.sample(n_samples)
    
    reduced_df = reduced_df.reset_index(drop=True)
    return reduced_df
```

---

El objetivo de este método es reducir la cantidad de columnas, esto lo hace combinando varias columnas en una.

Por ejemplo, un dataframe de 10 columnas se puede quedar en 5 columnas o en la que le indiquemos (siempre que sea menor al nº total de columnas)

Teniendo en cuenta que hay que buscar un equilibrio entre:
*   (+) columnas / (-) agrupamiento  -->  (-) precisión y (+) velocidad --- (puede sobreentrenar)
*   (-) columnas / (+) agrupamiento  -->  (+) precisión y (-) velocidad

El nº de columnas finales las definimos con la variable "n_components"
